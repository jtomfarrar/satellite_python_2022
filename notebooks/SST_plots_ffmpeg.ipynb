# SST Plots/create movie using ffmpeg

```python
import numpy as np
import xarray as xr
from matplotlib import pyplot as plt
import pandas as pd
from glob import glob
from dateutil import parser
import cartopy.crs as ccrs      
import cartopy #import projections
import gsw
import cftime
import datetime
```

```python
path='/Users/lfp/Desktop/SST_amsr2/2016/*.nc'
```

```python
fnames = glob(path)
fnames.sort()
#ds = xr.open_mfdataset(path,concat_dim='time',combine='nested')
for fname in fnames:
    print(parser.parse(fname.split('/')[-1].split('_')[-2]))
    
```

```python
ds = xr.open_mfdataset(fnames,concat_dim='time',combine='nested')
ds
```

```python
ds = ds.assign_coords(time = [parser.parse(fname.split('/')[-1].split('_')[-2]) for fname in fnames])
ds
```

```python
%matplotlib inline
#%matplotlib qt5

plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 400
plt.close('all')

__figdir__ = '/Users/lfp/Desktop/File1/'
savefig_args = {'bbox_inches':'tight', 'pad_inches':0.2}
plotfiletype='png'

# To save the plotted image
savefig = 1
```

```python
ds = ds.sel(lon=slice(80,100),lat=slice(5,25))
```

```python
for i in range(len(ds.time)):
    
    plt.close('all')

    ax = plt.axes(projection = ccrs.PlateCarree(central_longitude=200))  # Orthographic
        #day_str=ds.time.dt.strftime("%b %d %Y").values[i]
    ax.set_title(f'BoB SST {str(ds.isel(time=i).time)[-10:]}')
        #ax.set_title(f'BoB SST '+day_str)


    plt.set_cmap(cmap=plt.get_cmap('nipy_spectral'))
    plt.set_cmap(cmap=plt.get_cmap('turbo'))

    gl = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False, alpha=0.5, linestyle='--')
    gl.top_labels = False
    gl.ylabels_right = False

    levels = 10

    cs = ax.pcolormesh(ds.lon,ds.lat,np.squeeze(ds.isel(time=i).SST), vmin=25,vmax=32, transform=ccrs.PlateCarree())
    #ax.pcolormesh(ds.lon,ds.lat,np.squeeze(ds.SST), vmin=25,vmax=32, transform=ccrs.PlateCarree())

    cb = plt.colorbar(cs,fraction = 0.022, extend='both') 
    cb.set_label('SST [C]',fontsize = 10)
    ax.coastlines()
    ax.add_feature(cartopy.feature.LAND, zorder=3, facecolor=[.6,.6,.6], edgecolor='black')

       

    plt.savefig(__figdir__+f'SST{i:03d}'+'.'+plotfiletype,**savefig_args)
    plt.clf()
```

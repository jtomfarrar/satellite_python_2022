```python
ds = xr.open_mfdataset(path,concat_dim='time',combine='nested')
ds = ds.assign_coords(time = [parser.parse(fname.split('/')[-1].split('_')[-2]) for fname in glob(path)])
#ds = ds.assign_coords(time = [parser.parse(fname.split('/')[-1].split('.')[1]) for fname in glob(path)])
ds = ds.sortby('time')
ds
```

```python
ds.time
```

```python
#matplotlib inline
%matplotlib qt5

plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 400
plt.close('all')

__figdir__ = '../plots/plots3/'
savefig_args = {'bbox_inches':'tight', 'pad_inches':0.2}
plotfiletype='png'

# To save the plotted image
savefig = 1
```

```python
ds = ds.sel(lon=slice(80,100),lat=slice(5,25))
```

```python
for i in range(len(ds.time)):
    
    plt.close('all')
    
    ax = plt.axes(projection = ccrs.PlateCarree(central_longitude=200))  # Orthographic
    ax.set_title(f'BoB SST {str(ds.isel(time=i).time)[-10:]}')
    
    
    plt.set_cmap(cmap=plt.get_cmap('nipy_spectral'))
    plt.set_cmap(cmap=plt.get_cmap('turbo'))
    
    gl = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False, alpha=0.5, linestyle='--')
    gl.top_labels = False
    gl.ylabels_right = False

    levels = 10
    
    cs = ax.pcolormesh(ds.lon,ds.lat,np.squeeze(ds.isel(time=i).SST), vmin=25,vmax=32, transform=ccrs.PlateCarree())
    
    cb = plt.colorbar(cs,fraction = 0.022, extend='both') 
    cb.set_label('SST [C]',fontsize = 10)
    ax.coastlines()
    ax.add_feature(cartopy.feature.LAND, zorder=3, facecolor=[.6,.6,.6], edgecolor='black')
 
    #ax.set_extent([80,100,5,25]) #Set a specific location [xmin, xmax, ymin, ymax]
  
    plt.savefig(__figdir__+'SST_check2'+str(i)+'.'+plotfiletype,**savefig_args)
    #plt.clf()
```

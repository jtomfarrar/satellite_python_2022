```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def my_function(a,b,label,**kwarg):
    """
    (This is a header to explain the function does.)
    Plot a quadratic curve, y=ax+bx**2
    
    Inputs:
    a, b
    
    Outputs:
    None
    """
    
    x = np.arange(0,100,.1)
    y = a*x + b*x**2
    ax = plt.plot(x,y,**kwarg)
    plt.title(label)
    
    
    
```

```python
a = 2
b = 3
x = np.arange(0,100,.1)
y = a*x + b*x**2
ax = plt.plot(x,y)
```

```python
label = 'a=1, b=2'
my_function(1,2,label,color='green')
# my_function(20,2,color='red',linewidth=2)
```

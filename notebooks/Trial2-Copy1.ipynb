```python
#define the function and its derivative
def sinc(x):
    return np.sin(x)/x

def d_sinc(x):
    return np.cos(x)/x - np.sin(x)/x**2
```

```python
# Evaluate it at a certain range/points
n = 1000
x = np.linspace(-15, 15, n)
dx = x[1]-x[0]
f = sinc(x) 
df = d_sinc(x)
dfnum = np.gradient(f,dx)
fdif = dfnum - df
```

```python
# Plot the function and its derivative 
fig, ax = plt.subplots(2) 
#fig, ax = plt.subplots(2, figsize = (5, 10))
fig.tight_layout(pad=2.0) # space between the graphs
fig.set_size_inches(5, 10) # size of the graph/ left to right --- up to bottom 

ax[0].plot(x, f, color="red", label=r"$sinc(x)$")
ax[0].plot(x, df, color="blue", ls="--", label=r"$\frac{d(sinc(x))}{dx}$")
ax[0].plot(x, dfnum, color="green", ls="--", label=r"numdfdx")
ax[1].plot(x[1:n-1], fdif[1:n-1], color="purple", ls="--", label=r"error")


ax[0].set_title("Example Notebook Plot")
ax[0].set_xlabel(r"$x$ [radians]")
ax[1].set_title("Numerical - Analytical derivative")

ax[0].grid(True)
ax[1].grid(True)
ax[0].legend(loc='upper right');
ax[1].legend();
```
